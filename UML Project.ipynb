# KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
expected_clustors = range(1,11)
wcss = []
for i in expected_clustors:
    kmeans = KMeans(i)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14,4))
plt.plot(expected_clustors,wcss,marker='o')
plt.xlabel('Clustors')
plt.ylabel('WCSS/Inertia')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# PCA for dimensionality reduction
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df)
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Fit KMeans on the reduced data
kmeans = KMeans(n_clusters=4)
kmeans.fit(principal_df)
labels = kmeans.labels_

# Visualizing clusters
plt.figure(figsize=(10, 6))
scatter = plt.scatter(principal_df['PC1'], principal_df['PC2'], c=labels, cmap='viridis')
plt.title('KMeans Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Add legend
legend1 = plt.legend(*scatter.legend_elements(), title="Clusters")
plt.gca().add_artist(legend1)

plt.show()
```

```python
pca = PCA().fit(df)
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Explained Variance')
plt.show()
```

```python
# prompt: give code for silhouette score

from sklearn.metrics import silhouette_score

# Calculate silhouette score for the KMeans clustering
silhouette_score = silhouette_score(principal_df, labels)

# Print the silhouette score
print("Silhouette score:", silhouette_score)
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN

# Create the DBSCAN model
dbscan = DBSCAN(eps=10, min_samples=50) 

# Fit the model
dbscan.fit(principal_df)

# Visualizing clusters
plt.figure(figsize=(10, 6))
plt.scatter(principal_df['PC1'], principal_df['PC2'], c=dbscan.labels_, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, adjusted_rand_score

# Silhouette score
silhouette_avg = silhouette_score(principal_df, dbscan.labels_)
print("Silhouette Score:", silhouette_avg)
```

# One-Class SVM

```python
from sklearn.svm import OneClassSVM

# Create the One-Class SVM model
one_class_svm = OneClassSVM(nu=0.1)  # Adjust nu parameter as per your data

# Fit the model
one_class_svm.fit(X_train)

# Predict outliers/anomalies
outliers_svm = one_class_svm.predict(X_test)

# Anomaly scores
anomaly_scores_svm = one_class_svm.decision_function(X_test)

# Calculate the number of instances flagged as anomalies
num_predicted_anomalies_svm = np.sum(outliers_svm == -1)

# Calculate the expected number of anomalies based on the nu parameter
expected_num_anomalies_svm = int(len(X_test) * 0.1)  # Assuming nu = 0.1

# Compute accuracy
accuracy_svm = (len(X_test) - num_predicted_anomalies_svm) / len(X_test)

print("Number of instances flagged as anomalies:", num_predicted_anomalies_svm)
print("Expected number of anomalies based on nu parameter:", expected_num_anomalies_svm)
print("Accuracy of One-Class SVM model:", accuracy_svm)
```

```python
# One-Class SVM Anomaly Visualization
plt.figure(figsize=(10, 6))
plt.scatter(X_test.iloc[:, 0], X_test.iloc[:, 1], c=outliers_svm, cmap='viridis')
plt.title('One-Class SVM Anomaly Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Isolation Forest

```python
from sklearn.ensemble import IsolationForest

# Create the Isolation Forest model
isolation_forest = IsolationForest(contamination=0.1)  # Contamination represents the expected proportion of outliers in the data

# Fit the model
isolation_forest.fit(principal_df)

# Predict outliers/anomalies
outliers = isolation_forest.predict(principal_df)

# Anomaly scores
anomaly_scores = isolation_forest.decision_function(principal_df)
```

```python
anomaly_scores
```

```python
import numpy as np

# Calculate the number of instances flagged as anomalies
num_predicted_anomalies = np.sum(outliers == -1)

# Calculate the expected number of anomalies based on the contamination rate
expected_num_anomalies = int(len(principal_df) * 0.1)

# Compute accuracy
accuracy = (len(principal_df) - num_predicted_anomalies) / len(principal_df)

print("Number of instances flagged as anomalies:", num_predicted_anomalies)
print("Expected number of anomalies based on contamination rate:", expected_num_anomalies)
print("Accuracy of Isolation Forest model:", accuracy)
```

```python
import matplotlib.pyplot as plt

# Visualize anomalies detected by Isolation Forest
plt.figure(figsize=(10, 6))
plt.scatter(principal_df['PC1'], principal_df['PC2'], c=outliers, cmap='viridis')
plt.title('Isolation Forest Anomaly Detection')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Anomaly Score')
plt.show()
```
